```
from scipy.interpolate import RBFInterpolator

def load_feature_points(csv_path):
    """Load feature points from CSV file."""
    df = pd.read_csv(csv_path)
    if csv_path not in non_flipped:
        df['x1'] = 511 - df['x1']
        df['y1'] = 511 - df['y1']
    source_points = df[['x2', 'y2']].values  # Feature positions in image 1
    target_points = df[['x1', 'y1']].values  # Corresponding positions in image 2
    return source_points, target_points

def warp_image(image, source_points, target_points, scale_factor=1.0):
    """Warp image using thin-plate spline transformation."""
    h, w = image.shape[:2]
    
    # Scale the source points
    source_points_scaled = source_points * scale_factor
    
    # Create a grid of pixel coordinates
    grid_x, grid_y = np.meshgrid(np.arange(w), np.arange(h))
    grid_points = np.column_stack([grid_x.ravel(), grid_y.ravel()])
    
    # Interpolate transformation using RBF (thin-plate spline kernel)
    rbf_x = RBFInterpolator(source_points_scaled, target_points[:, 0], kernel='thin_plate_spline')
    rbf_y = RBFInterpolator(source_points_scaled, target_points[:, 1], kernel='thin_plate_spline')
    
    # Compute the transformed coordinates
    warped_x = rbf_x(grid_points).reshape(h, w).astype(np.float32)
    warped_y = rbf_y(grid_points).reshape(h, w).astype(np.float32)
    
    # Remap the image
    warped_image = cv2.remap(image, warped_x, warped_y, interpolation=cv2.INTER_CUBIC)
    
    return warped_image, rbf_x, rbf_y

def add_scale_bar(ax, length, pixel_length, location=(0.1, 0.9), color='white', linewidth=2):
    """Add a scale bar to an image."""
    ax.plot([location[0], location[0] + (1/pixel_length)], [location[1], location[1]], color=color, linewidth=linewidth, transform=ax.transAxes)
    ax.text(location[0] + (1/pixel_length) / 2, location[1] - 0.05, f'{length} µm', color=color, ha='center', va='top', transform=ax.transAxes)

def main(afm_file, cl_file, image1, image2, csv_path, output_path, scale_factor=1.0):
    """Main function to align image1 to image2."""

    
    # Load feature points
    source_points, target_points = load_feature_points(csv_path)
    
    # Warp image1 to align with image2
    warped_image1, rbf_x, rbf_y = warp_image(image1, source_points, target_points, scale_factor)
    
    # Save and display results
    # cv2.imwrite(output_path, warped_image1)

    # # Uncomment the following lines to visualize the results

    plt.figure(figsize=(11, 11))
    
    ax1 = plt.subplot(2, 2, 1)
    ax1.imshow(image2, cmap='viridis')
    ax1.scatter(source_points[:,0], source_points[:,1], facecolors='none', edgecolors="red", label="Source")
    ax1.set_title(os.path.basename(cl_file))
    add_scale_bar(ax1, 1, 4)  # 1 µm scale bar for 4 µm image
    
    ax2 = plt.subplot(2, 2, 2)
    ax2.imshow(image1, cmap='afmhot')
    ax2.scatter(target_points[:,0], target_points[:,1], facecolors='none', edgecolors="red", label="Target")
    ax2.set_title(os.path.basename(afm_file))
    add_scale_bar(ax2, 1, 3)  # 1 µm scale bar for 3 µm image
    
    ax3 = plt.subplot(2, 2, 3)
    ax3.imshow(warped_image1, cmap='afmhot')
    ax3.set_title("AFM Image (Aligned)")
    ax3.set_xlim(0, 256)
    ax3.set_ylim(256, 0)  # Invert y-axis to match image coordinates
    add_scale_bar(ax3, 1, 4)  # 1 µm scale bar for 3 µm image
    
    ax4 = plt.subplot(2, 2, 4)
    ax4.imshow(image2, cmap='viridis')
    ax4.imshow(warped_image1, cmap='afmhot', alpha=0.3)  # Overlay with transparency
    ax4.set_title("Overlay of Aligned AFM Image on CL Image")
    ax4.set_xlim(0, 256)
    ax4.set_ylim(256, 0)  # Invert y-axis to match image coordinates
    add_scale_bar(ax4, 1, 4)  # 1 µm scale bar for 4 µm image
    
    # plt.savefig(output_path.replace('.png', '_comparison.png'), dpi=300)
    plt.show()

    return image1, image2, warped_image1, rbf_x, rbf_y



all_data_dict = {}
for afm_file, cl_file, alignment_file in zip(afm_files_processed, cl_files, alignment_files):
    all_data_dict[os.path.basename(afm_file)] = {
        'afm_file': afm_file,
        'cl_file': cl_file,
        'alignment_file': alignment_file
    }

    # Load images
    # afm_img = cv2.imread(afm_file, cv2.IMREAD_GRAYSCALE)
    afm_img = np.load(afm_file)  # Load pre-processed AFM numpy array. It no longer needs to be flipped horizontally
    cl_sem = hs.load(cl_file, signal_type='CL_SEM')
    im = cl_sem.T


    # Uncomment the following lines to visualize the CL image and ROI before image extraction
    # im.plot()
    # roi1 = hs.roi.SpanROI(left=350, right=380) #sets a digitalbandfilter
    # im_roi1 = roi1.interactive(im, color="red")
    # im_roi1_mean = hs.interactive(im_roi1.mean,
    #                         event=roi1.events.changed,
    #                         recompute_out_event=None)

    # cl_img = im_roi1_mean.data

    # Apply bandpass filter numerically (rather than plotting)
    # Extract energy axis and get index range for 350–380 nm
    energy_axis = im.axes_manager[0].axis  # Assumes axis 0 is spectral
    mask = (energy_axis >= 350) & (energy_axis <= 380)

    # Extract mean over bandpass range
    cl_img = im.data[mask].mean(axis=0)



    # Perform Gaussian Multifit on CL Image

    # cl_sem_eV_si = cl_sem.to_eV(inplace=False, jacobian=False)
    # m_si = cl_sem_eV_si.create_model()
    # bkg_si = hs.model.components1D.Offset()
    # g1_si = hs.model.components1D.GaussianHF()
    # m_si.extend([g1_si, bkg_si])
    # m_si.components
    # cl_sem_eV_si.axes_manager.indices = (50,50)
    # g1_si.centre.value = 3.4        # Gaussian centre
    # g1_si.fwhm.value = 0.1      # Gaussian width
    # g1_si.height.value = 500      # Gaussian height
    # bkg_si.offset.value = 200   # background offset
    # g1_si.centre.bmax = g1_si.centre.value + 0.2
    # g1_si.centre.bmin = g1_si.centre.value - 0.2
    # g1_si.fwhm.bmin = 0.01
    # m_si.fit()
    # m_si.assign_current_values_to_all()
    # m_si.plot(plot_components=True)
    # m_si.multifit(bounded=True, show_progressbar=True)
    # m_si.plot(plot_components=True)

    # m_centre = g1_si.centre.as_signal()
    # m_intensity = g1_si.height.as_signal()
    # m_fwhm = g1_si.fwhm.as_signal()

    # m_centre.save(os.path.splitext(cl_file)[0] + "_m_centre" + os.path.splitext(cl_file)[1])
    # m_intensity.save(os.path.splitext(cl_file)[0] + "_m_intensity" + os.path.splitext(cl_file)[1])
    # m_fwhm.save(os.path.splitext(cl_file)[0] + "_m_fwhm" + os.path.splitext(cl_file)[1])


    # Load previously saved model results if available

    # m_centre = hs.load(os.path.splitext(cl_file)[0] + "_m_centre" + os.path.splitext(cl_file)[1])
    # m_intensity = hs.load(os.path.splitext(cl_file)[0] + "_m_intensity" + os.path.splitext(cl_file)[1])
    # m_fwhm = hs.load(os.path.splitext(cl_file)[0] + "_m_fwhm" + os.path.splitext(cl_file)[1])



    # Define output path for the aligned image
    aligned_img_output_path = os.path.join(os.path.dirname(afm_file), "aligned_" + os.path.basename(afm_file))

    image1, image2, warped_image1, rbf_x, rbf_y = main(afm_file, cl_file, afm_img, cl_img, alignment_file, aligned_img_output_path, scale_factor=1.0)

    all_data_dict[os.path.basename(afm_file)]['aligned_image'] = warped_image1
    all_data_dict[os.path.basename(afm_file)]['rbf_x'] = rbf_x
    all_data_dict[os.path.basename(afm_file)]['rbf_y'] = rbf_y
    all_data_dict[os.path.basename(afm_file)]['afm_img'] = image1
    all_data_dict[os.path.basename(afm_file)]['cl_img'] = image2
    all_data_dict[os.path.basename(afm_file)]['cl_raw'] = cl_sem.data
    # all_data_dict[os.path.basename(afm_file)]['m_centre'] = m_centre
    # all_data_dict[os.path.basename(afm_file)]['m_intensity'] = m_intensity
    # all_data_dict[os.path.basename(afm_file)]['m_fwhm'] = m_fwhm

    


# if __name__ == "__main__":
#     image1 = afm_img  # Variable with AFM image
#     image2 = im_roi1_mean.data  # Variable with CL image
#     csv_path = click_filename  # Path to CSV file with feature mappings
#     output_path = aligned_img_output_path  # Output file path
    
#     image1, image2, warped_image1 = main(image1, image2, csv_path, output_path, scale_factor=1.0)  # Adjust scale_factor as needed
```

```
from skimage import io
from skimage import filters, measure
from scipy import ndimage as ndi
import matplotlib.pyplot as plt
import numpy as np

spots_all = []
spot_sizes_all = []
spot_depths_all = []

min_spot_size = 7
max_spot_size = 100


# Loop through all images in afm_files
for idx, file in enumerate(afm_files):
    # Read the image
    # image = cv2.imread(file, cv2.IMREAD_GRAYSCALE)

    image = io.imread(file, as_gray=True)
    if image.min() >= 0.0 and image.max() <= 1.0: #convert from float image to 16 bit image if needed
        image = (image * 65535).astype(np.uint16)

    # Fit a plane to the image
    x, y = np.meshgrid(np.arange(image.shape[1]), np.arange(image.shape[0]))
    A = np.c_[x.ravel(), y.ravel(), np.ones_like(x.ravel())]
    coeff, _, _, _ = np.linalg.lstsq(A, image.ravel(), rcond=None)
    plane = (coeff[0] * x + coeff[1] * y + coeff[2]).reshape(image.shape)

    # Set any pixel with a value higher than the plane to the level of the plane
    filtered_image = np.minimum(image, plane)

    # Apply Gaussian blur to reduce noise
    # blurred_image = filters.gaussian(image, sigma=1)

    # Perform local thresholding with adjusted parameters
    threshold_image = filters.threshold_local(filtered_image, block_size=19, offset=3072)
    dark_spots_image = filtered_image < threshold_image

    # Label the dark spots
    labeled_spots, num_spots = ndi.label(dark_spots_image)

    # Find the coordinates and sizes of the spots
    spots = measure.regionprops(labeled_spots)
    spot_sizes = [spot.area for spot in spots if min_spot_size <= spot.area <= max_spot_size]
    spots_rightsize = [spot for spot in spots if min_spot_size <= spot.area <= max_spot_size]
    spots_all.append(spots_rightsize)
    all_data_dict[os.path.basename(file)]['spots'] = spots_rightsize
    all_data_dict[os.path.basename(file)]['spot_sizes'] = spot_sizes
    spot_sizes_all.append(spot_sizes)

    # Plot the image with the spots highlighted
    fig, ax = plt.subplots(1, 2, figsize=(12, 6))

    pixel_size_um = 3.0 / image.shape[0]

    ax[0].imshow(image, cmap='afmhot', extent=[0, 3, 3, 0])
    ax[0].set_title(f'Image {idx + 1}')

    ax[0].set_xlim([0, 3])
    ax[0].set_ylim([3, 0])  # Flip y-axis so origin is top-left
    ax[0].set_xlabel('x (μm)')
    ax[0].set_ylabel('y (μm)')

    ax[1].imshow(image, cmap='afmhot', extent=[0, 3, 3, 0])
    for spot in spots:
        if 5 <= spot.area <= 100:
            cy, cx = spot.centroid
            contour = measure.find_contours(labeled_spots == spot.label, 0.5)
            for n, contour in enumerate(contour):
                # Convert contour pixel coords to µm
                x_um = contour[:, 1] * pixel_size_um
                y_um = contour[:, 0] * pixel_size_um
                ax[1].plot(x_um, y_um, linewidth=1, color='blue')

    ax[1].set_title(f'Image {idx + 1} with Dark Spots Highlighted via Local Thresholding')
    ax[1].set_xlim([0, 3])
    ax[1].set_ylim([3, 0])  # Flip y-axis so origin is top-left
    ax[1].set_xlabel('x (μm)')
    ax[1].set_ylabel('y (μm)')

    plt.show()


    # # Plot a range of block_size and offset values
    # block_sizes = [15, 31, 51, 71]
    # offsets = [5, 10, 15, 20]

    # fig, axes = plt.subplots(len(block_sizes), len(offsets), figsize=(15, 15))
    # for i, block_size in enumerate(block_sizes):
    #     for j, offset in enumerate(offsets):
    #         threshold_image = filters.threshold_local(image, block_size=block_size, offset=offset)
    #         dark_spots_image = image < threshold_image
    #         labeled_spots, num_spots = ndi.label(dark_spots_image)
    #         spots = measure.regionprops(labeled_spots)
    #         spot_sizes = [spot.area for spot in spots if 6 <= spot.area <= 100]

    #         axes[i, j].imshow(image, cmap='afmhot')

    #         for spot in spots:
    #             if 6 <= spot.area <= 100:
    #                 cy, cx = spot.centroid
    #                 contour = measure.find_contours(labeled_spots == spot.label, 0.5)
    #                 for n, contour in enumerate(contour):
    #                     axes[i, j].plot(contour[:, 1], contour[:, 0], linewidth=1, color='blue')

    #         axes[i, j].set_title(f'block_size={block_size}, offset={offset}')
    #         axes[i, j].axis('off')

    # plt.tight_layout()
    # plt.show()

    # Calculate the mean intensity (depth) of each spot relative to the surrounding pixels
    spot_depths = []
    for spot in spots:
        if min_spot_size <= spot.area <= max_spot_size:
            # Get the coordinates of the spot
            coords = spot.coords
            
            # Create a mask for the spot
            spot_mask = np.zeros_like(image, dtype=bool)
            spot_mask[coords[:, 0], coords[:, 1]] = True
            
            # Create a mask for the surrounding pixels
            surrounding_mask = ndi.binary_dilation(spot_mask) & ~spot_mask
            
            # Calculate the mean intensity of the spot and the surrounding pixels
            spot_mean = np.mean(image[spot_mask])
            surrounding_mean = np.mean(image[surrounding_mask])
            
            # Calculate the depth as the difference between the spot and the surrounding pixels
            spot_depth = spot_mean - surrounding_mean
            spot_depths.append(spot_depth)

    print(f"Number of spots in image {idx + 1}: {len(spot_sizes)}")
    spot_depths_all.append(spot_depths)
    all_data_dict[os.path.basename(file)]['spot_depths'] = spot_depths

    # Plot the histograms of spot sizes and depths
    plt.figure(figsize=(10, 7))

    plt.subplot(1, 2, 1)
    plt.hist(spot_sizes, bins=30, color='blue', alpha=0.7)
    plt.title(f'Histogram of Spot Sizes in Image {idx + 1}')
    plt.xlabel('Spot Size (pixels)')
    plt.ylabel('Frequency')

    plt.subplot(1, 2, 2)
    plt.hist(spot_depths, bins=30, color='green', alpha=0.7)
    plt.title(f'Histogram of Spot Depths in Image {idx + 1}')
    plt.xlabel('Spot Depth (mean intensity)')
    plt.ylabel('Frequency')

    plt.tight_layout()
    plt.show()

    # if idx == 5:
    #     break

```
